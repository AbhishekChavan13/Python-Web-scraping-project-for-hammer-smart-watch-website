```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL='https://hammeronline.in/collections/smart-watch?sort_by=title-ascending'
URL
```

```python
r = requests.get(URL)
htmlcontent = r.content
htmlcontent
```

```python
soup = BeautifulSoup(htmlcontent,"html.parser")
print(soup)
soup.prettify
```

```python
names=soup.find_all(class_="grid-product__title grid-product__title--body")
print(names)
```

```python
Smartwatches_name=[]
for i in range(0,len(names)):
    Smartwatches_name.append(names[i].get_text())
print(Smartwatches_name)
```

```python
original_price=soup.find_all(class_="grid-product__price--original")
print(original_price)
```

```python
Smartwatches_price=[]
for i in range(0,len(original_price)):
    Smartwatches_price.append(original_price[i].get_text())
print(Smartwatches_price)
```

```python
reviews=soup.find_all(class_="jdgm-prev-badge")
print(reviews)
```

```python
reviews_given=[]
for i in range(0,len(reviews)):
    reviews_given.append(reviews[i].get_text())
print(reviews_given)
```

```python
detail=soup.find_all(class_="collection_tags")
print(detail)
```

```python
products_detail=[]
product_detail=[]
for i in range(0,len(detail)):
    products_detail.append(detail[i].get_text())
#print(product_detail)
for i in range(0,len(detail)):
    product_detail.append(products_detail[i].strip())
print(product_detail)
```

```python
Discounted_price=soup.find_all(class_="grid-product__price")
Discounted_price
```

```python
Discounted_Price=[]
discounted_price=[]
for i in range(0,len(Discounted_price)):
    Discounted_Price.append(Discounted_price[i].get_text())
#print(Discounted_Price)
for i in range(0,len(Discounted_price)):
    discounted_price.append(Discounted_Price[i].strip())
#print(discounted_price)
Discounted_Price=[i[:+12] for i in discounted_price]
print(Discounted_Price)
```

```python
hammer_Smartwatches=pd.DataFrame({"Smartwatches_name":Smartwatches_name,"product_detail":product_detail,"reviews_count":reviews_given,"original_price":Smartwatches_price,"discounted_price":Discounted_Price},index=range(1,len(Smartwatches_name)+1))
print(hammer_Smartwatches)
```
